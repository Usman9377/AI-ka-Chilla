### Composition
- What are the different categories or groups in the data?
- How many unique values are there for categorical variables?
- What is the proportion of each category in the dataset?

```python
# Import Library
import pandas as pd
```

```python
# load the data
df = pd.read_csv("titanic.csv") 
df.head() # this shows glimpse of data
```

```python
df.shape
# print number of rows and columns using f string
print(f"The dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
df.info() # this shows data types and non-null counts
```

```python
df.dtypes # this shows data types of each column
```

```python
df["sex"].value_counts()  # this shows count of unique values in
```

```python
# find value count percentage for sex column
df["sex"].value_counts() / len(df) * 100 # type: ignore
# round to 2 decimal places with formula round(numbers, decimal places)
(df["sex"].value_counts() / len(df) * 100).round(2)
```

```python
(df["class"].value_counts() / len(df) * 100).round(2) # for class column 
# sort values in asscending order
(df["class"].value_counts() / len(df) * 100).round(2).sort_values()
```

```python
# find the value counts in percentage for class where sex==female
df_females = df[df["sex"]=="female"]
df_females["class"].value_counts() / len(df_females) * 100
```

```python
# group by sex and class to get value counts in percentage
df.groupby("sex")["class"].value_counts(normalize=True) * 100
# round 2 decimal
```
